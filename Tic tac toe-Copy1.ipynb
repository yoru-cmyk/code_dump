```python




#: board,int ----> int,int
def human_place_mark(board,playing):
    
    while True: 
        pos = int(input("Please enter position (Acceptable choices: 0-8)"))
        if pos in [i for i in range(0,9)]:
            break
        else:
            print(pos)
            print("Hey human, I said acceptable choices were between 0 and 8.  Try Again.")
         
    new_board = place_mark(board,playing,pos)
    
    if new_board != None:
        return new_board
    else:
        print("Nice try human. But it seems that spot is taken. Try again")
        return human_place_mark(board,playing)


# board,int,int --- > board
def place_mark(board,playing,pos):
    if playing == 0:
        mark = X
    elif playing == 1:
        mark = O    
    
    
    if board[pos] == 0:
        new_board=board.copy()
        new_board[pos] = mark
        
        return new_board
    else:
        return None
        
    
#:: int,board,winner ---> int,board 
def play_turn(n_turns,tree):
    
    board=tree.board
    
    print(str(board[0]) + " | " + str(board[1]) + " | " + str(board[2]))
    print(str(board[3]) + " | " + str(board[4]) + " | " + str(board[5]))
    print(str(board[6]) + " | " + str(board[7]) + " | " + str(board[8]))

    if check_winning(board) == 1:
        print("Player 0 victory")
        return 1
    #Flatten the board and then check if there are 0 (empty spaces)
    elif check_winning(board) == -1:
        print("Player 1 victory")
        return -1
    elif check_winning(board) == 0:
        print("It's a tie")
        return 0
    else:
        print("Keep playing!")
    
    playing = n_turns % 2 
    
    if playing == 0:
        print(f"Player 0 turn")
        
        if ai_player_0 == True:
            new_tree = tree.ai_choose()
        else:
            board = human_place_mark(board,playing)
            new_tree = [kid for kid in tree.kids if kid.board == board][0]
    elif playing == 1:
        print(f"Player 1 turn")
        
        if ai_player_1 == True:
            new_tree = tree.ai_choose()
        else:
            board = human_place_mark(board,playing)
            new_tree = [kid for kid in tree.kids if kid.board == board][0]

    play_turn(n_turns+1,new_tree)


#::board ---> Int/winner or None 
def check_winning(board):
    all_combos = (#horizontal combinations
                       (0,1,2),
                      (3,4,5),
                      (6,7,8),
                        #Vertical combos
                      (0,3,6),
                      (1,4,7),
                      (2,5,8),
                        #diagonal combos
                      (0,4,8),
                      (6,4,2))
    
    #Sums up cases across lines to check if there is a win for either 0 or 1
    
    maybe_winning_sums = [(board[x[0]] + board[x[1]] + board[x[2]]) for x in all_combos]

    if 3 in maybe_winning_sums:
        return 1
    elif 15 in maybe_winning_sums:
        return -1
    elif 0 not in board:
        return 0
    else:
        return None
    
n_turns = 0


X = 1
O = 5

class Tree:
    def __init__(self,board,n_turns,name):
        self.name = name
        self.board = board
        self.empty = [i for i in range(len(board)) if board[i] == 0]
        self.playing = n_turns % 2

        # 0 = tie, 1= player who plays first wins, -1 = player who plays second wins
        self.score = check_winning(self.board)
        #self.marks = [place_mark(board,self.playing,pos) for pos in self.empty]
        self.kids = None if self.score != None else [Tree(place_mark(board,self.playing,x),n_turns+1,name + str(x)) for x in self.empty]
        self.score = self.morph_kid()
        
    #Assign score to self based on kids
    def morph_kid(self):
        if self.kids != None:
            kid_scores = [kid.score for kid in self.kids]
            #If every kid is a terminal node
            if None not in kid_scores:
                return max(kid_scores) if self.playing == 0 else min(kid_scores)
            else:
                return self.score
        else:
            return self.score
        
    def ai_choose(self):
        kid_scores = [kid.score for kid in self.kids]
        if self.playing == 0:
            best_index = kid_scores.index(max(kid_scores))
            return self.kids[best_index]
        elif self.playing ==1:
            worst_index = kid_scores.index(min(kid_scores))
            return self.kids[worst_index]


name1 = input("Please enter your name player 0 (enter 'AI' for an AI):")

name2 = input("Please enter your name player 1 (enter 'AI' for an AI):")

ai_player_0 = True if name1=="AI" else False
ai_player_1 = True if name2=="AI" else False

print("----------!LOADING in progress!------------")

# Source for ASCII art:https://animeeverything.online/anime-dot-arts-copy-and-paste-and-how-to-make-them/#Some_popular_Anime_Braille_ASCII_artsForYouTube
board = [0,0,0,0,0,0,0,0,0]
tree = Tree(board,0,"")


print("-----------------LOADING DONE!---------------------")
print("Players start your engines!")

result = play_turn(0,tree)
    
print("-------------GAME OVER!-----------------")
print("Thanks for playing")

print("""⢸⣿⣿⣿⣿⠃⠄⢀⣴⡾⠃⠄⠄⠄⠄⠄⠈⠺⠟⠛⠛⠛⠛⠻⢿⣿⣿⣿⣿⣶⣤⡀⠄
⢸⣿⣿⣿⡟⢀⣴⣿⡿⠁⠄⠄⠄⠄⠄⠄⠄⠄⠄⠄⠄⠄⠄⠄⣸⣿⣿⣿⣿⣿⣿⣿⣷
⢸⣿⣿⠟⣴⣿⡿⡟⡼⢹⣷⢲⡶⣖⣾⣶⢄⠄⠄⠄⠄⠄⢀⣼⣿⢿⣿⣿⣿⣿⣿⣿⣿
⢸⣿⢫⣾⣿⡟⣾⡸⢠⡿⢳⡿⠍⣼⣿⢏⣿⣷⢄⡀⠄⢠⣾⢻⣿⣸⣿⣿⣿⣿⣿⣿⣿
⡿⣡⣿⣿⡟⡼⡁⠁⣰⠂⡾⠉⢨⣿⠃⣿⡿⠍⣾⣟⢤⣿⢇⣿⢇⣿⣿⢿⣿⣿⣿⣿⣿
⣱⣿⣿⡟⡐⣰⣧⡷⣿⣴⣧⣤⣼⣯⢸⡿⠁⣰⠟⢀⣼⠏⣲⠏⢸⣿⡟⣿⣿⣿⣿⣿⣿
⣿⣿⡟⠁⠄⠟⣁⠄⢡⣿⣿⣿⣿⣿⣿⣦⣼⢟⢀⡼⠃⡹⠃⡀⢸⡿⢸⣿⣿⣿⣿⣿⡟
⣿⣿⠃⠄⢀⣾⠋⠓⢰⣿⣿⣿⣿⣿⣿⠿⣿⣿⣾⣅⢔⣕⡇⡇⡼⢁⣿⣿⣿⣿⣿⣿⢣
⣿⡟⠄⠄⣾⣇⠷⣢⣿⣿⣿⣿⣿⣿⣿⣭⣀⡈⠙⢿⣿⣿⡇⡧⢁⣾⣿⣿⣿⣿⣿⢏⣾
⣿⡇⠄⣼⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⠟⢻⠇⠄⠄⢿⣿⡇⢡⣾⣿⣿⣿⣿⣿⣏⣼⣿
⣿⣷⢰⣿⣿⣾⣿⣿⣿⣿⣿⣿⣿⣿⣿⢰⣧⣀⡄⢀⠘⡿⣰⣿⣿⣿⣿⣿⣿⠟⣼⣿⣿
⢹⣿⢸⣿⣿⠟⠻⢿⣿⣿⣿⣿⣿⣿⣿⣶⣭⣉⣤⣿⢈⣼⣿⣿⣿⣿⣿⣿⠏⣾⣹⣿⣿
⢸⠇⡜⣿⡟⠄⠄⠄⠈⠙⣿⣿⣿⣿⣿⣿⣿⣿⠟⣱⣻⣿⣿⣿⣿⣿⠟⠁⢳⠃⣿⣿⣿
⠄⣰⡗⠹⣿⣄⠄⠄⠄⢀⣿⣿⣿⣿⣿⣿⠟⣅⣥⣿⣿⣿⣿⠿⠋⠄⠄⣾⡌⢠⣿⡿⠃
⠜⠋⢠⣷⢻⣿⣿⣶⣾⣿⣿⣿⣿⠿⣛⣥⣾⣿⠿⠟⠛⠉⠄⠄""")
    
```

```python
def cracklepop(n):
    if n == 101:
        print("END!")
        return 0
    
    elif n % 5 == 0 and n % 3 == 0:
        print("cracklepop")
        return cracklepop(n+1)
    elif n % 3 == 0:
        print("crackle")
        return cracklepop(n+1)
    elif n % 5 == 0:
        print("pop")
        return cracklepop(n+1)
    else:
        print(n)
        return cracklepop(n+1)

cracklepop(1)
```
